```python
import time
import sys

sys.set_int_max_str_digits(50000)

def recur_factorial(n):
    if n == 1 or n == 0:
        return 1
    else:
        return n * recur_factorial(n - 1)

def iter_factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

try:
    num = int(input("Enter a non-negative number: "))
    print ("nnumber entered :",num,"\n")
    if num < 0:
        print("Factorial does not exist for negative numbers")
    else:
        start_time = time.time()
        iter_res = iter_factorial(num)
        iter_time = time.time() - start_time
        
        start_time = time.time()
        recur_res = recur_factorial(num)
        recur_time = time.time() - start_time
        
        print(f"Iterative Execution Time: {iter_time:.10f} seconds")
         
        print("\n calculation.......\n")


        print(f"Recursive Execution Time: {recur_time:.10f} seconds")
        print("\n calculated value:",recur_res)
        
except ValueError:
    print("Invalid input!")
```

```python
import math

def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
        
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

def generate_primes(limit):
    primes = []
    for i in range(2, limit + 1):
        if is_prime(i):
            primes.append(i)
    return primes

def next_prime(n):
    current = n + 1
    while True:
        if is_prime(current):
            return current
        current += 1

def find_twin_primes(limit):
    twins = []
    primes = generate_primes(limit)
    for i in range(len(primes) - 1):
        if primes[i + 1] - primes[i] == 2:
            twins.append((primes[i], primes[i + 1]))
    return twins

try:
    limit = int(input("Enter a limit for prime generation and twin primes: "))
    num = int(input("Enter a number to find the next prime after it: "))

    print(f"\nPrimes up to {limit}:")
    print(generate_primes(limit))

    print(f"\nNext prime after {num}:")
    print(next_prime(num))

    print(f"\nTwin primes up to {limit}:")
    print(find_twin_primes(limit))
    
except ValueError:
    print("Invalid input!")
```
