```python
## list opration 
# build in functions 

lst=[9,43,3,4,5,4,5,6,32,1]
lst.sort()
print(lst)
```

```python
#[start pos :end pos: diffrences ] ---> normal case 
#[start from back pos:upto lower pos : diff from back -int]----> reverse  traversing 

print(lst2[::])
print(lst2[1::])
print(lst2[1:8:])
print(lst2[0:8:2])
print(lst2[::-1])
print(lst2[::-2])
print(lst2[7:4:-1])
```
