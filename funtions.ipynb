```python
# parameterize funs
# passing the values

# made a prime num fun

def isnumprime(num):
    if num <= 1:
        print("not a prime number:", num)
        return
    for i in range(2, num):
        if num % i == 0:
            print("not a prime number:", num)
            return
    print("prime number:", num)


isnumprime(6)
```
