```python
def sumnum (a,b):
     return a+b


def subnum (a,b):
     return a-b


def mulnum (a,b):
     return a*b


def divnum (a,b):
     return a/b
               
```

```python
a=int(input( "enter number"))
b=int(input( "enter number"))

print (" numbers entered : ",a ,b)

print ("sum is :",sumnum(a,b))
print ("sum is :",subnum(a,b))
print ("sum is :",mulnum(a,b))
print ("sum is :",divnum(a,b))
```
